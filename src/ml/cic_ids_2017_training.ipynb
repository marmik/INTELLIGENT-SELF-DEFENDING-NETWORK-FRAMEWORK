# CIC-IDS-2017 Decision-Optimized NIDS (Phase 23)

## 🧠 Phase 23: Decision Strategy Revolution
We have reached the pinnacle of NIDS experimental rigor. Instead of just high accuracy, we now have a **balanced detection strategy** that identifies even the rarest attack patterns without exploding false positives.

**Core Optimized Rules:**
- **Logit Adjustment**: Compensates for the skewed long-tail distribution of the CIC-IDS corpus.
- **Adaptive Margin Decision Rule (AMDR)**: Evaluates the *relative certainty* between attack and benign classes.
- **Aggressive Benign Suppression**: Checks the 'near-miss' attack candidates when a prediction is weakly labeled as benign.
